```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
import io
from google.colab import files
upload = files.upload()
df=pd.read_csv(io.BytesIO(upload['StudentsPerformance.csv']))
print(df)
df.shape
df.describe()
df.plot(x='test preparation course', y='math score', style='o')  
plt.title('test preparation course vs math score')  
plt.xlabel('test preparation course')  
plt.ylabel('math score')  
plt.show()
plt.figure(figsize=(6,4))
plt.tight_layout()
seabornInstance.distplot(df['math score'])
plt.figure(figsize=(6,4))
plt.tight_layout()
seabornInstance.distplot(df['reading score'])
plt.figure(figsize=(6,4))
plt.tight_layout()
seabornInstance.distplot(df['writing score'])
X = df['reading score'].values.reshape(-1,1)
y = df['writing score'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
#To retrieve the intercept:
print(regressor.intercept_)

#For retrieving the slope:
print(regressor.coef_)
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
print(df)
df1 = df.head(25)
df1.plot(kind='bar',figsize=(10,8))
plt.title('Actual vs Predicted')
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
