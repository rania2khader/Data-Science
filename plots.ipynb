```python
import numpy as np
import matplotlib.pyplot as plt

height = np.random.normal(170, 6, 100)
weight = (height-100)*np.random.uniform(0.75, 1.25, 100)
plt.title("Height and Weight distribution of 100 Persons", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Height", fontdict={'fontsize': 10, 'fontweight' : 3, 'color' : 'Blue'})
plt.ylabel("Weight", fontdict={'fontsize': 10, 'fontweight' : 3, 'color' : 'Blue'})  
plt.scatter(height, weight, color = 'red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

year = np.array([1920,1930,1940,1950,1960,1970,1980,1990,2000,2010])
unemployment = np.random.normal(9.0,2.7,10)
y = list(range(1920,2020,10))
plt.title("Unemployment rate over the last 10 decades")
plt.xlabel("Year")
plt.ylabel("Unemployment")
plt.xticks(y)
plt.plot(year, unemployment)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
   
np.random.seed(10) 
laptops = np.random.normal(100, 20, 1000) 
desktops = np.random.normal(200, 30, 1000)
tab = np.random.uniform(150,10,1000)
fig = plt.figure(figsize=(10,7))
plt.boxplot([laptops, desktops, tab]) 
plt.title("Sales of products in States")
plt.ylabel("Sales")
plt.xticks((1,2,3),["Laptops","Desktops","Tab"])

plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

years = np.array(['2012', '2013', '2014', '2015', '2016'])

sales_Africa = np.random.uniform(10,1,5)
sales_Asia_Pacific = np.random.uniform(40,1,5)
sales_Europe = np.random.uniform(80,1,5)
fig, ax = plt.subplots()

africa, = ax.plot(years, sales_Africa)
africa.set_label("Africa")

aspac, = ax.plot(years, sales_Asia_Pacific)
aspac.set_label("Asia Pacific")

europe, = ax.plot(years, sales_Europe)
europe.set_label("Europe")

plt.show()
```
