```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


pstore = pd.read_csv("weather1.csv")
pstore.head()
plt.style.use('seaborn-white')
#Create a distribution plot for mean_temp  
sns.distplot(pstore.mean_temp  )
plt.show()

#Change the number of bins
sns.distplot(pstore.mean_temp, bins=20, kde = False)
plt.show()



#Create a distribution plot for rating
sns.distplot(pstore.mean_temp, bins=20, color="g")
plt.title("Distribution of app ratings", fontsize=20, color = 'red')
plt.show()

pstore = pstore[~pstore['rainfall'].isin([0,'Trace',.1,.2,.3])]
pstore['rainfall'].value_counts()

plt.figure(figsize=[9,7])
pstore['rainfall'].value_counts().plot.pie()
plt.show()


sns.jointplot(pstore.RH, pstore.mean_temp, kind = "reg")
plt.show()
```
